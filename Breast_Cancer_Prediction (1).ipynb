```python
a=pd.get_dummies(data["diagnosis"])
```

```python
a
```

```python
cancer=pd.concat([data,a],axis="columns")
```

```python
cancer.head()
```

```python
cancer.drop(["diagnosis","B"],axis="columns",inplace=True)
```

```python
cancer
```

```python
cancer.rename(columns={"M":"Malignant/Benign"},inplace=True)
```

```python
cancer
```

```python
y=cancer[["Malignant/Benign"]]
```

```python
y
```

```python
X=cancer.drop(["Malignant/Benign"],axis="columns")
```

```python
X
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=np.array(X)
y=np.array(y)
```

```python
type(X)
```

```python
type(y)
```

Experiment 1:Logistic Regression :-

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(X_train,y_train)
```

```python
y_pred=logreg.predict(X_test)
```

```python
y_pred
```

```python
score=metrics.accuracy_score(y_pred,y_test)
```

```python
score
```

```python
print("The accuracy of prediction is {} %".format(score*100))
```

Experiment 2:KNN 

```python
knn=KNeighborsClassifier(n_neighbors=24)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
```

```python
y_pred
```

```python
score=metrics.accuracy_score(y_test,y_pred)
```

```python
score
```

```python
print("The accuracy of prediction is {} %".format(score*100))
```

Experiment 3:Decision Tree:-

```python
clf=tree.DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
score=metrics.accuracy_score(y_test,y_pred)
```

```python
score
```

```python
print("The accuracy of prediction is {} %".format(score*100))
```
